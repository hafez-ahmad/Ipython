```python
import mpl_toolkits.basemap as bm
mapproj = bm.Basemap(projection='cyl',llcrnrlat=-20.0, llcrnrlon=-180.0,urcrnrlat=20.0, urcrnrlon=180.0)
mapproj.drawcoastlines()
mapproj.drawparallels(np.array([-20, -10, 0, 10, 20]),labels=[1,0,0,0])
mapproj.drawmeridians(np.array([-180, -90, 0, 90, 180]),labels=[0,0,0,1])
lonproj, latproj = mapproj(lonall, latall)
plt.contour(lonproj, latproj, tt)
```
